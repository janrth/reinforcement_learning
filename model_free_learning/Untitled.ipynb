```python
import gym

import temporal_differencing_on_policy
```

```python
# instantiate class

env = gym.make('FrozenLake-v1')

td = temporal_differencing_on_policy.TDOnPolicy(env=env,
                 num_episodes=100000, 
                 num_timesteps=1000,
                 epsilon=.95,
                 alpha=.8,
                 gamma=.85)
```

```python
# run temporal differencing on-policy to update Q values:

Q = td.policy_computation()
```

```python
# extract optimal pol

df = td.find_optimal_policy(Q)
```
