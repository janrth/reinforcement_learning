```python
from models import monte_carlo_method
import gym
```

```python
# instantiate env and monte-carlo prediction:

env = gym.make('Blackjack-v1')

mc = monte_carlo_method.MonteCarloPrediction(env=env, 
                                             num_timesteps=1000,
                                             num_iterations=100000)
```

```python
# run monte carlo prediction:

df = mc.value_function()
```

```python
# Look at average return (value) for each state:

df.sort_values('value',ascending=False)
```
